##4.1.4 設定圖表及xy軸標題

```python
import matplotlib.pyplot as plt

listx=[1,5,7,9,13]
listy=[15,50,80,40,600]

plt.plot(listx,listy,color="red",lw="2.0",ls="--",label="Label")
plt.legend()
plt.title("Chart title",fontsize=20) #圖表標題
plt.xlabel("X label",fontsize=14) #y軸標題
plt.ylabel("Y label",fontsize=14) #x軸標題
plt.show()
```

##4.1.5 設定xy軸資料範圍

```python
import matplotlib.pyplot as plt
listx=[1,5,7,9,13,16]
listy=[15,50,80,40,70,50]
plt.plot(listx,listy,color="red",lw="2.0",ls="--",label="Label")
plt.legend()
plt.title("Chart title",fontsize=20)
plt.xlabel("X label",fontsize=14)
plt.ylabel("Y label",fontsize=14)
plt.xlim(0,20) #設定x軸範圍
plt.ylim(0,100) #設定y軸範圍
plt.show()
```

##4.1.6 設定格線

```python
import matplotlib.pyplot as plt
listx=[1,5,7,9,13,16]
listy=[15,50,80,40,70,50]
plt.plot(listx,listy,color="red",lw="2.0",ls="--",label="Label")
plt.legend()
plt.title("Chart title",fontsize=20)
plt.xlabel("X label",fontsize=14)
plt.ylabel("Y label",fontsize=14)
plt.xlim(0,20)
plt.ylim(0,100)
plt.grid(color="black",ls=':',lw=1,alpha=0.5) #設定格線
#alpha=0.5 的 alpha 參數表示透明度（opacity）
plt.show()
```

##4.1.8 自訂軸刻度

```python
import matplotlib.pyplot as plt
listx=[1000,2000,3000,4000,5020]
listy=[15,50,80,40,70]
plt.plot(listx,listy)
plt.xticks(listx) #設定或取得 x 軸的刻度位置與標籤
#注意看那個5020
plt.tick_params(axis='both',labelsize=16,color="red")#axis也可以用 'x' 或 'y' 只改其中一軸。
plt.show()
```

##4.1.9範例 : 個年度銷售表

```python
import matplotlib.pyplot as plt

year=[2015,2016,2017,2018,2019]
city1=[128,150,199,180,150]
plt.plot(year,city1,'r--s',lw=2,ms=10,label="Taipei")

city2=[120,145,180,170,120]
plt.plot(year,city2,'g-.*',lw=2,ms=10,label="Taichung")
plt.legend()

plt.ylim(50,250)
plt.xticks(year)
plt.title("Sales Report",fontsize=18)
plt.xlabel("Year",fontsize=12)
plt.ylabel("Million",fontsize=12)
plt.grid(color='k',ls=':',lw=1,alpha=0.5)
plt.show()
```

#4.3 圓形圖pie

```python
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.font_manager import fontManager
fontManager.addfont('/content/SourceHanSansTW/SubsetOTF/TW/SourceHanSansTW-Regular.otf')
matplotlib.rc('font',family='Source Han Sans TW')

sizes=[25,30,15,10]
labels=["北部","西部","南部","中部"]
colors=["red","green","blue","yellow"]
explode=(0,0,0.2,0)
plt.pie(sizes,
  labels=labels,
  colors=colors,
  explode=explode,
  labeldistance=1.1,
  autopct="%2.1f%%",
  pctdistance=0.6,
  shadow=True,
  startangle=90)
plt.show()
```

#4.7 在圖表區加入多張圖表

##4.7.1 用欄列排列多張圖表:subplot()

```python
import matplotlib.pyplot as plt
plt.figure(figsize=[8,8])
plt.subplot(2,1,1) #有點不懂其中三個參數的意義
plt.title(label='Chart 1',fontsize=20)
plt.plot([1,2,3],'r:o')

plt.subplot(2,1,2)
plt.title(label='Chart 2',fontsize=20)
plt.plot([1,2,3],'g--o')
plt.show()
```

##4.7.2 用相對位置排列多張圖表:axes

```python
import matplotlib.pyplot as plt
plt.figure(figsize=[8,8])
plt.axes([0,0,0.4,1]) #小心，傳入裡面的必須要是一個list
plt.title(label='Chart 1',fontsize=20)
plt.plot([1,2,3],'r:o')

plt.axes([0.5,0,0.4,1]) #與左邊界距離
plt.title(label='Chart 2',fontsize=20)
plt.plot([1,2,3],'g--o')

plt.show()
```

```python
import matplotlib.pyplot as plt
plt.axes([0,0,0.8,1])
plt.title(label='Chart 1',fontsize=20)
plt.plot([1,2,3],'r:o')

plt.axes([0.55,0.1,0.2,0.2])
plt.title(label='Chart 2',fontsize=20)
plt.plot([1,2,3],'g--o')
```
